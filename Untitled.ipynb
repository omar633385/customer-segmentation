# K_means clustering

```python
n_clusters = range(2,13)
inertia_errors = []
silhouette_scores = []
for n in n_clusters:
    model=make_pipeline(
    StandardScaler(),
     KMeans(n_clusters=n,random_state=0)
    )
    model.fit(X)
    inertia_errors.append(model.named_steps['kmeans'].inertia_)
    silhouette_scores.append(silhouette_score(X, model.named_steps['kmeans'].labels_))
    



plt.plot(n_clusters,inertia_errors,'-x')
plt.xlabel('Number of Clusters , k')
plt.ylabel('inertia')
plt.xticks(n_clusters)
```

```python
final_model = make_pipeline(
StandardScaler(),
    KMeans(n_clusters=5,random_state=0)
)
final_model.fit(X)
```

```python
labels = final_model.named_steps['kmeans'].labels_
xgb = X.groupby(final_model.named_steps['kmeans'].labels_).mean()
xgb
```

```python
fig = px.bar(xgb,labels={'x': 'Cluster', 'y': 'Value [$]'},
             title='Mean of features for each Cluster') 
fig.show()
```

```python
def show_clusters(data, labels):
    
    plt.plot(figsize=(15,10))
    palette = sns.color_palette('hls', n_colors=len(set(labels)))
    sns.scatterplot(x=data.iloc[:, 0], y=data.iloc[:, 1], hue=labels, palette=palette)
    plt.show()

show_clusters(X, labels)
```

# DBscan

```python
from sklearn.cluster import DBSCAN

eps= np.arange(0.3,10.2,0.2)
min_samples = np.arange(2,11)
noise=[]
best_noise = float('inf')
best_score = float('-inf')
best_model = None
for i in eps:
    for j in min_samples:
        dbSCAN_Model = DBSCAN(eps=i, min_samples=j).fit(X)
        labels = dbSCAN_Model.labels_
        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        
        if  n_clusters  < 5 or n_clusters  >10 :
            continue
        
        noise_percentage = (list(labels).count(-1) / len(X)) * 100
        noise.append(noise_percentage)
        
        score = silhouette_score(X, labels)
        
        if (noise_percentage < best_noise) or (noise_percentage == best_noise and score > best_score):
            best_noise = noise_percentage
            best_score = score
            best_model = dbSCAN_Model
            
print(best_score)
print(best_noise)
best_model
```

```python
dbscan = DBSCAN(eps=9.100000000000001, min_samples=2)
db_scan_clusters = dbscan.fit_predict(X)
```

```python
show_clusters(X,db_scan_clusters)
```

# hierarchy clustering

```python
from scipy.cluster.hierarchy import linkage, dendrogram

Z = linkage(X, method='ward', metric='euclidean')


plt.figure(figsize=(10, 5))
dendrogram(Z)
plt.xlabel('Customers')
plt.ylabel('Distance')
plt.title('Hierarchical Clustering Dendrogram')
plt.show()
```
